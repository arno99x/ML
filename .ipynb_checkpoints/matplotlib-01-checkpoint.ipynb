## 散点图 scatter plot
### 散点与拆线的用法没差异，但是使用场景不同
#### 散点图：横纵两轴都是特征
#### 拆线图：横轴是特征，纵轴是取值
#### 所以一般双特征的数据用散点图

```python
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x,siny,label="sin(x)")
plt.scatter(x,cosy,color="red" ,label="sin(y)")
plt.axis([-1,11,-2,2])
plt.xlabel="x axis"
plt.ylabel="y axis"
plt.title("welcome to the ML World!")
plt.legend()
plt.show()
```

```python
# 绘制正态分布图， 中间部分较密，周围较稀疏
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)

plt.scatter(x,y)
plt.show()
```

```python
# alpha 透明度
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)

plt.scatter(x,y,alpha=0.5)
plt.show()
```
