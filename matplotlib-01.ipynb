```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.xlim(-5,15)
plt.ylim(-3,1.5)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.axis([-1,11,-2,2])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x,siny , label="sin(x)")
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)")
plt.axis([-1,11,-2,2])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x,siny , label="sin(x)")
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)")
plt.axis([-1,11,-2,2])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("welcome to the ML World!")
plt.show()
```
